## **Definições**

**Classes:** proporcionam uma forma de organizar dados e funcionalidades juntos. Criar uma nova classe cria um novo “tipo” de objeto, permitindo que novas “instâncias” desse tipo sejam produzidas.  

**Métodos:** Uma instância de uma classe, é como se fosse uma função porém pertencendo a uma classe específica. Modificam a classe de alguma forma.

**Módulo:** Um arquivo .py que contém conjunto de funções, classes reunidos num namespace comum.

--> **Exemplo:** ao salvar um arquivo como teste_modulo.py
podemos "chamá-lo" em outro arquivo usando suas funções, classes, etc, basta digitar
`import teste_modulo`

**Biblioteca:** Conjunto de módulos que estão juntos e que podem ser usados num programa ou numa outra biblioteca.

**Pacote:** Unidade de distribuição que pode conter uma biblioteca, um executável ou ambos. 


## Criando uma classe

```python
class Dna:
  def __init__(self, sequence):
    self.sequence = sequence
  
  def add_nucleotide(self, nucleotide):
    self.nucleotide = nucleotide
    return self.sequence + self.nucleotide
```

### Criando o objeto Dna

```python
objeto_dna = Dna("ATCAGTCAGCTAGCAT") #instanciando
type(objeto_dna)
```

```python
print(objeto_dna.sequence)
```

```python
objeto_dna.sequence = "ATC"
```

```python
print(objeto_dna.sequence)
```

### Chamando um método da classe

```python
print(objeto_dna.add_nucleotide("A"))
```

## Classe SeqFeature

É com essa classe que se cria o atributo feature do SeqRecord. A classe SeqFeature tenta encapsular o máximo possível de informações sobre a sequência. O design é fortemente baseado nas tabelas de recursos do GenBank / EMBL.

A ideia principal sobre cada objeto SeqFeature é descrever uma região em uma sequência específica. Essa região é descrita com um **objeto de localização**, normalmente um intervalo entre duas posições (mRna, gene, exons, introns, etc).

Atributos da classe SeqFeature:

1. **type**: descreve o tipo de feature, ex: gene, mRNA ...
2. **locations**: para descrever as regiões da sequência, usamos o atributo locations e outros métodos que veremos mais para frente.
3. **qualifiers**: é um dicionário Python de informações adicionais. ex: uma chave poderia ser ["evidência"] = "não experimental" para indicar que o gene foi encontrado de maneira computacional.
4. **sub_features**: IGNOREM ESSE ATRIBUTO. Se tornou obsoleto com a criação do objeto CompoundLocation (veremos a seguir).


```python
from Bio import SeqFeature
```

**Objeto FeatureLocation**

Em genes de procarioto, a maioria das localizações do SeqFeature são extremamente simples - você só precisa de coordenadas de início e fim e uma fita. Isso é essencialmente tudo o que o objeto FeatureLocation básico faz.

Em eucariotos, há altas chances de termos que lidar com locais compostos (óperons). Além disso, as próprias posições podem ser confusas (inexatas). Para isso, foi criado o **Objeto CompoundLocation**, que junta regiões de genes em locais diferentes.

Para analisar genes com posições não tão certas, temos o conceito de **posições fuzzy**, mas como é muito específico, não trouxe aqui. Para saber mais: [4.3.2.3 do Manual Python](https://biopython.org/DIST/docs/tutorial/Tutorial.html#sec%3Alocations/) 



Observe que CDS do GenBank ou arquivos EMBL é a união dos exons - eles não cobrem nenhum íntron.

## Descrevendo a feature de uma sequência sua

```python
from Bio.SeqFeature import SeqFeature, FeatureLocation

feature_1 = SeqFeature(FeatureLocation(5, 18), type="gene", strand=-1)
feature_2 = SeqFeature(FeatureLocation(20, 40), type="gene", strand=+1)
```

Com o método extract é possível extrair a sequência relacionado com aquela feature

## Acessando objetos SeqRecord pelo seu reverso complementar

**PS: é o metodo reverse_complement do objeto SeqRecord, é diferente do método reverse_complement do objeto Seq**

Ele inverte as sequências, se houver anotação específica para nucleotídeos, ele mantém, porém a transferência da maioria da informação é problemática. Portanto, por padrão, o id, nome, descrição, anotações e referências cruzadas do banco de dados SeqRecord não são transferidos.


```python
luci_reversec = record_luci.reverse_complement()
print(luci_reversec)
```

O método reverse_complement do objeto SeqRecord leva um número de argumentos opcionais correspondentes às propriedades do registro. **Definir esses argumentos como True significa copiar os valores antigos**, enquanto False significa descartar os valores antigos e usar o valor padrão. Como alternativa, você pode fornecer o novo valor desejado.

```python
luci_reversec = record_luci.reverse_complement(id=True)
print(luci_reversec)
```

```python
luci_reversec = record_luci.reverse_complement(id="A89SA8S9A")
print(luci_reversec)
```
