## **Definies**

**Classes:** proporcionam uma forma de organizar dados e funcionalidades juntos. Criar uma nova classe cria um novo tipo de objeto, permitindo que novas instncias desse tipo sejam produzidas.  

**Mtodos:** Uma instncia de uma classe,  como se fosse uma funo porm pertencendo a uma classe especfica. Modificam a classe de alguma forma.

**Mdulo:** Um arquivo .py que contm conjunto de funes, classes reunidos num namespace comum.

--> **Exemplo:** ao salvar um arquivo como teste_modulo.py
podemos "cham-lo" em outro arquivo usando suas funes, classes, etc, basta digitar
`import teste_modulo`

**Biblioteca:** Conjunto de mdulos que esto juntos e que podem ser usados num programa ou numa outra biblioteca.

**Pacote:** Unidade de distribuio que pode conter uma biblioteca, um executvel ou ambos. 


## Criando uma classe

```python
class Dna:
  def __init__(self, sequence):
    self.sequence = sequence
  
  def add_nucleotide(self, nucleotide):
    self.nucleotide = nucleotide
    return self.sequence + self.nucleotide
```

### Criando o objeto Dna

```python
objeto_dna = Dna("ATCAGTCAGCTAGCAT") #instanciando
type(objeto_dna)
```

```python
print(objeto_dna.sequence)
```

```python
objeto_dna.sequence = "ATC"
```

```python
print(objeto_dna.sequence)
```

### Chamando um mtodo da classe

```python
print(objeto_dna.add_nucleotide("A"))
```

## Classe SeqFeature

 com essa classe que se cria o atributo feature do SeqRecord. A classe SeqFeature tenta encapsular o mximo possvel de informaes sobre a sequncia. O design  fortemente baseado nas tabelas de recursos do GenBank / EMBL.

A ideia principal sobre cada objeto SeqFeature  descrever uma regio em uma sequncia especfica. Essa regio  descrita com um **objeto de localizao**, normalmente um intervalo entre duas posies (mRna, gene, exons, introns, etc).

Atributos da classe SeqFeature:

1. **type**: descreve o tipo de feature, ex: gene, mRNA ...
2. **locations**: para descrever as regies da sequncia, usamos o atributo locations e outros mtodos que veremos mais para frente.
3. **qualifiers**:  um dicionrio Python de informaes adicionais. ex: uma chave poderia ser ["evidncia"] = "no experimental" para indicar que o gene foi encontrado de maneira computacional.
4. **sub_features**: IGNOREM ESSE ATRIBUTO. Se tornou obsoleto com a criao do objeto CompoundLocation (veremos a seguir).


```python
from Bio import SeqFeature
```

**Objeto FeatureLocation**

Em genes de procarioto, a maioria das localizaes do SeqFeature so extremamente simples - voc s precisa de coordenadas de incio e fim e uma fita. Isso  essencialmente tudo o que o objeto FeatureLocation bsico faz.

Em eucariotos, h altas chances de termos que lidar com locais compostos (perons). Alm disso, as prprias posies podem ser confusas (inexatas). Para isso, foi criado o **Objeto CompoundLocation**, que junta regies de genes em locais diferentes.

Para analisar genes com posies no to certas, temos o conceito de **posies fuzzy**, mas como  muito especfico, no trouxe aqui. Para saber mais: [4.3.2.3 do Manual Python](https://biopython.org/DIST/docs/tutorial/Tutorial.html#sec%3Alocations/)

## Acessando objetos SeqRecord pelo seu reverso complementar

**PS:  o metodo reverse_complement do objeto SeqRecord,  diferente do mtodo reverse_complement do objeto Seq**

Ele inverte as sequncias, se houver anotao especfica para nucleotdeos, ele mantm, porm a transferncia da maioria da informao  problemtica. Portanto, por padro, o id, nome, descrio, anotaes e referncias cruzadas do banco de dados SeqRecord no so transferidos.


```python
luci_reversec = record_luci.reverse_complement()
print(luci_reversec)
```

O mtodo reverse_complement do objeto SeqRecord leva um nmero de argumentos opcionais correspondentes s propriedades do registro. **Definir esses argumentos como True significa copiar os valores antigos**, enquanto False significa descartar os valores antigos e usar o valor padro. Como alternativa, voc pode fornecer o novo valor desejado.

```python
luci_reversec = record_luci.reverse_complement(id=True)
print(luci_reversec)
```

```python
luci_reversec = record_luci.reverse_complement(id="A89SA8S9A")
print(luci_reversec)
```
