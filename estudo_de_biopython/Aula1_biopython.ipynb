**Resposta:**

- Muta√ß√µes no TATA Box podem levar a:
  - Diminui√ß√£o da transcri√ß√£o (menos express√£o da prote√≠na correspondente);
  - Aus√™ncia da transcri√ß√£o esperada.

- **Por exemplo:** Se numa amostra sequenciada de um paciente espera-se encontrar o TATA Box na regi√£o promotora do gene de interesse, podemos suspeitar de um processo de muta√ß√£o o qual est√° associado a algumas doen√ßas em humanos que podem ser:

## **ORF vs. CDS**

**IMPORTANTE:**
- Uma **ORF (*Open Reading Frame*)** √© uma sequ√™ncia de DNA que come√ßa com o c√≥don de in√≠cio ‚ÄúATG‚Äù (nem sempre) e termina com qualquer um dos tr√™s c√≥dons de termina√ß√£o (TAA, TAG, TGA). A ORF √© geralmente prevista com base na sequ√™ncia de DNA e n√£o foi comprovada a transcri√ß√£o;
- Dependendo do ponto de partida, existem 6 maneiras poss√≠veis (3 na fita direta ou parental e 3 na fita complementar) de traduzir qualquer sequ√™ncia de nucleot√≠deos em sequ√™ncia de amino√°cidos de acordo com o c√≥digo gen√©tico. Essas s√£o chamadas de **frames de leitura**.
- Uma **CDS (*Coding Sequence*)** √© a regi√£o real do DNA que √© traduzida para formar prote√≠nas (sem os √≠ntros, no caso de eucariotos), em outras palavras uma CDS se refere √†queles nucleot√≠deos (exons concatenados) que podem ser divididos em c√≥dons que s√£o traduzidos em amino√°cidos pelo mecanismo de tradu√ß√£o ribossomal.

## **Start-codon**

**IMPORTANTE:**
- O c√≥don de in√≠cio √© o primeiro c√≥don de um transcrito de RNA mensageiro (mRNA) traduzido por um ribossomo;
- O c√≥don inicial sempre codifica para Metionina (Met) em eucariotos e Archaea e uma N-formilmetionina (fMet) em bact√©rias, mitoc√¥ndrias e plast√≠deos;
- O c√≥don de in√≠cio mais comum √© AUG (isto √©, ATG na sequ√™ncia de DNA correspondente);
- Os c√≥dons de in√≠cio alternativos s√£o diferentes do c√≥don AUG padr√£o e s√£o encontrados tanto em procariotos (bact√©rias e arqu√©ias) quanto em eucariotos;
- Os c√≥dons de in√≠cio alternativo ainda s√£o traduzidos como Met quando est√£o no in√≠cio de uma prote√≠na (mesmo se o c√≥don codificar um amino√°cido diferente de outra forma).

**Eucariotos:**
- C√≥dons de in√≠cio alternativos (n√£o-AUG) s√£o muito raros em genomas eucari√≥ticos;
- C√≥dons de in√≠cio n√£o AUG de ocorr√™ncia natural foram relatados para alguns mRNAs celulares;
- Sete das nove poss√≠veis substitui√ß√µes de nucleot√≠deo √∫nico no c√≥don inicial AUG da diidrofolato redutase foram funcionais como locais de in√≠cio da tradu√ß√£o em c√©lulas de mam√≠feros;
- Al√©m da via can√¥nica do c√≥don Met-tRNA Met e AUG, as c√©lulas de mam√≠feros podem iniciar a tradu√ß√£o com valina usando um valil-tRNA espec√≠fico que decodifica o c√≥don CUG.
- *Candida albicans* usa um c√≥don de in√≠cio CAG.

**Procariotos:**
- Os procariotos usam c√≥dons de in√≠cio alternativo, principalmente GUG e UUG.
- 83% dos CDCs de *E. coli* usam AUG como c√≥don de in√≠cio, 14% usam GUG e 3% usam AUU e CUG. Quando algum destes c√≥dons est√° sendo usado como c√≥don de in√≠cio, a tradu√ß√£o sempre ser√° para Metionina, embora usualmente seja traduzido para outro amino√°cido, exemplo: AUU normalmente traduz para uma Isoleucina, mas como c√≥don de in√≠cio n√£o-estandar, traduz para uma Metionina;
- As regi√µes de codifica√ß√£o bem conhecidas que n√£o t√™m c√≥dons de inicia√ß√£o AUG s√£o aquelas de lacI (GUG) e lacA (UUG) no operon lac de *E. coli*.

**Mitoc√¥ndrias:**
- Os genomas mitocondriais usam c√≥dons de in√≠cio alternativos de forma mais significativa (AUA e AUU em humanos).

## **Mudando os nucleot√≠deos de uma sequ√™ncia**

**IMPORTANTE:**
- As letras de uma sequ√™ncia de um objeto Seq s√£o imut√°veis:

````python
# Criando a vari√°vel...
my_seq = Seq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
my_seq
````

````python
# Tentando trocar uma letra da sequ√™ncia anterior...
my_seq[5] = "G"
````

Por√©m, isto pode mudar, usando o seguinte m√©todo:

````python
# Trocando uma letra da sequ√™ncia anterior sem obter erro...
from Bio.Seq import MutableSeq
mutable_seq = MutableSeq("GCCATTGTAATGGGCCGCTGAAAGGGTGCCCGA")
mutable_seq
````

````python
# Trocando um "C" por um "" na quinta posi√ß√£o da sequ√™ncia anterior...
mutable_seq[5] = "C"
mutable_seq
````

````python
# Removendo o primeiro "T" localizado na quarta posi√ß√£o...
mutable_seq.remove("T")
mutable_seq
````

Sequ√™ncia modificada 2: GCCACGTAATGGGCCGCTGAAAGGGTGCCCGA

````python
# Invertendo a sequ√™ncia anterior...
mutable_seq.reverse()
mutable_seq
````

````python
# Voltando a criar um objeto "Seq"...
new_seq = mutable_seq.toseq()
new_seq
````

## **Nucleot√≠deos desconhecidos de uma sequ√™ncia**

**Problema Biol√≥gico:**

- √â comum obter letras desconhecidas ap√≥s sequenciamento de DNA ou prote√≠nas.
- Para sequ√™ncias de DNA ou RNA, nucleot√≠deos desconhecidos s√£o comumente denotados pela letra ‚ÄúN‚Äù, enquanto que para prote√≠nas, ‚ÄúX‚Äù √© comumente usado para amino√°cidos desconhecidos. 
- O objeto UnknownSeq √© uma subclasse do objeto Seq b√°sico e seu objetivo √© representar uma sequ√™ncia em que descobramos o seu tamanho, mas n√£o as letras reais que a comp√µem. 
- Pode ser usado um objeto Seq normal nesta situa√ß√£o, mas gasta muita mem√≥ria para armazenar uma string de um milh√£o de caracteres ‚ÄúN‚Äù quando voc√™ poderia simplesmente armazenar uma √∫nica letra ‚ÄúN‚Äù e o comprimento desejado como um inteiro .
- Alguns formatos de arquivo de sequ√™ncia nem sempre incluem a sequ√™ncia real, por exemplo, os arquivos GenBank e EMBL podem incluir uma lista de recursos, mas para a sequ√™ncia apenas apresenta as informa√ß√µes contig.

**Formatos de arquivos:**

**FASTA vs. QUAL:**

- Os **arquivos QUAL** usados no trabalho de sequenciamento possuem pontua√ß√µes de qualidade, mas nunca cont√™m uma sequ√™ncia - em vez disso, est√° o **arquivo FASTA** que possui a sequ√™ncia;
- O formato de arquivo QUAL √© um formato adicional relacionado ao FASTA.
Um arquivo FASTA √†s vezes √© acompanhado por um arquivo QUAL, principalmente quando o arquivo fasta cont√©m sequ√™ncias geradas em um instrumento de sequenciamento de alto rendimento;
- Os arquivos QUAL armazenam uma pontua√ß√£o de qualidade Phred (inteiro n√£o negativo) para cada base em uma sequ√™ncia armazenada no formato FASTA.
